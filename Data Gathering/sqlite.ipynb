```python
import sqlite3
```

```python
## Connect to an sqlite database
connection = sqlite3.connect('example.db')
connection
```

```python
cursor = connection.cursor()
```

```python
## Create a Table
cursor.execute('''
Create Table if not exists employees(
    id Integer Primary Key,
    name Text Not Null,
    age Integer,
    department text
    )
''')
## Commit the changes
connection.commit()
```

```python
cursor.execute('''
Select * from employees
''')
```

```python
## Inset the data in sqlite table
cursor.execute("""
INSERT INTO employees(name, age, department)
VALUES ('Priyanshu yadav', 21, 'Data Scientist')
""")

cursor.execute("""
INSERT INTO employees(name, age, department)
VALUES ('Shudhanshu yadav', 21, 'software developer')
""")

cursor.execute("""
INSERT INTO employees(name, age, department)
VALUES 
('nimesh', 21, 'cyber security'),
('gaurav kumar', 21, 'cyber security'),
('dipesh', 21, 'Data Scientist')
""")
# commit the change
connection.commit()
```

-	execute() → database se data maangta hai
-	fetchall() → saara data le aata hai
-	rows → data list ke form me store hota hai

#### fetchall() kya karta hai?

Jab aap database se data SELECT karte ho, to data cursor ke paas ready hota hai.
fetchall() us saare data ko ek saath nikal leta hai.


```python
# Query the data from the table
cursor.execute('Select * from employees')
rows = cursor.fetchall()
# cursor.fetchone()  --> sirf ek row

# print the queried data
for row in rows:
    print(row)
```

```python
# Update the data in the table
cursor.execute('''
UPDATE employees
       Set age = 20
       where name = "Priyanshu yadav" 
''')
connection.commit()
```

```python
cursor.execute('Select * from employees')
rows = cursor.fetchall()
for row in rows:
    print(row)
```

```python
# Delete the data from table
cursor.execute('''
Delete from employees
               where name = 'dipesh'
''')
connection.commit()
```

```python
cursor.execute('Select * from employees')
rows = cursor.fetchall()
for row in rows:
    print(row)
```

```python
# Working with Sales Data
# Connect ta an Sqlite database
connection = sqlite3.connect('sales_data.db')
cursor = connection.cursor()

# Create a tavle for sales data
cursor.execute('''
CREATE TABLE IF NOT EXISTS sales (
     id INTEGER PRIMARY KEY,
     data TEXT NOT NULL,
     product TEXT NOT NULL,
     sales INTEGER,
     region TEXT
)
''')

# Insert data into the sales table
sales_data = [
    ('2023-01-01', 'Product1', 100, 'North'),
    ('2023-01-01', 'Product2', 200, 'South'),
    ('2023-01-01', 'Product3', 150, 'North'),
    ('2023-01-01', 'Product1', 250, 'East'),
    ('2023-01-01', 'Product5', 104, 'West')
]
cursor.executemany('''
insert into sales(data,product,sales,region)
                   values(?,?,?,?)
''',sales_data)
connection.commit()
```

```python
cursor.execute('Select * from sales')
rows = cursor.fetchall()
for row in rows:
    print(row)
```

```python
# Close the connection
connection.close()
```
