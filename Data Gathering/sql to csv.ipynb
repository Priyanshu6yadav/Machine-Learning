## Convert directly sql to csv

```python
import pandas as pd
import re

rows = []

with open("world.sql", encoding="latin1") as f:
    for line in f:
        if line.startswith("INSERT INTO `city`"):
            values = re.findall(r"\((.*?)\)", line)
            for v in values:
                rows.append([x.strip("'") for x in v.split(",")])

# create dataframe
df = pd.DataFrame(rows, columns=["ID","Name","CountryCode","District","Population"])

df.to_csv("city.csv", index=False)

print("CSV created successfully!")
```
