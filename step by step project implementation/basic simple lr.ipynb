```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('height-weight (1).csv')
df = pd.DataFrame(df)
```

```python
df.head()
```

```python
plt.scatter(df['Weight'],df['Height'])
plt.xlabel("Weight")
plt.ylabel("Height")
```

```python
X = df[["Weight"]]
y = df[["Height"]]  # [[]] Make Height and weight in 2 dimentions
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.20,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train) ## fit() computes mean and standard deviation (not median).
X_test=scaler.transform(X_test)      ## transform() applies the Z-score formula using those statistics.
```

```python
plt.scatter(X_train,y_train)
```

```python
### train the simple linear regression model
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
```

```python
regressor.fit(X_train,y_train)
```

```python
print("The slop or cofficient of weight is", regressor.coef_)
print("Intercept: ",regressor.intercept_)
```

```python
## Applying straight line formula hQ(x)= 0o + 01x1
# 0o = 157.5
# 01 = 17.03440872
plt.scatter(X_train,y_train)
plt.plot(X_train,regressor.predict(X_train),color ='g')
```

#### Predictions of train data
1. prediction height output = intercept +coef_(Weight)
2. y_pred_train = 157.5 + 17.03(X_train)

#### Prediction of test data
1. Prediction height output = intercept + coef_(Weights)
2. y_pred_test = 157.5 + 17.03(x_test)

```python
y_pred_test = regressor.predict(X_test)
```

```python
y_pred_test,y_test
```

```python
plt.scatter(X_test,y_test)
plt.plot(X_test,regressor.predict(X_test),color ='g')
```

### Performance Metrics
MSE,MAE,RMSE 
R square and adjusted R square

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
mse = mean_squared_error(y_test,y_pred_test)
mae = mean_absolute_error(y_test,y_pred_test)
rmse = np.sqrt(mse)
```

```python
print(mse)
print(mae)
print(rmse)
```

```python
# Finding accuracy of model
from sklearn.metrics import r2_score
```

```python
score=r2_score(y_test,y_pred_test)
```
