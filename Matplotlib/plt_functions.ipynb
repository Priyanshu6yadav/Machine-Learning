```python
## Import necessary libraries
import matplotlib.pyplot as plt  # Main plotting library
import numpy as np               # For numerical operations
```

```python
# 📊 1. BAR CHART

# Data for bar chart
x = ['A', 'B', 'C', 'D']        # Categories
y = [10, 15, 7, 12]             # Values

# Basic syntax:
# plt.bar(x_values, y_values, color=..., width=..., label=...)
plt.bar(x, y, color='skyblue')
plt.title('Vertical Bar Chart')         # Title of chart
plt.xlabel('Categories')                # X-axis label
plt.ylabel('Values')                    # Y-axis label
plt.show()                              # Display the plot
```

```python
## 📊 2. HORIZONTAL BAR CHART


# Basic syntax:
# plt.barh(y_values, x_values, color=...)
plt.barh(x, y, color='orange')
plt.title('Horizontal Bar Chart')
plt.xlabel('Values')
plt.ylabel('Categories')
plt.show()
```

```python
# 📉 3. HISTOGRAM

# Data for histogram
data = np.random.randn(1000)  # 1000 random numbers from normal distribution

# Basic syntax:
# plt.hist(data, bins=..., color=..., edgecolor=...)
plt.hist(data, bins=20, color='purple', edgecolor='black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# 🥧 4. PIE CHART

# Data for pie chart
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [30, 25, 20, 25]
colors = ['red', 'yellow', 'pink', 'brown']
explode = (0.1, 0, 0, 0)  # Explode the first slice

# Basic syntax:
# plt.pie(sizes, labels=..., colors=..., explode=..., autopct=..., shadow=..., startangle=...)
plt.pie(sizes, labels=labels, colors=colors, explode=explode, autopct='%1.1f%%', shadow=True, startangle=90)
plt.title('Fruit Distribution')
plt.axis('equal')  # Keep pie chart circular
plt.show()
```

```python
#🧭 5. VECTOR PLOT (QUIVER PLOT)

# Create a grid of coordinates
X, Y = np.meshgrid(np.arange(0, 5, 1), np.arange(0, 5, 1))

# Define vector components
U = np.cos(X)  # X-direction vectors
V = np.sin(Y)  # Y-direction vectors

# Basic syntax:
# plt.quiver(X, Y, U, V)
plt.quiver(X, Y, U, V)
plt.title('Vector Field (Quiver Plot)')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# 📈 6. LINE PLOT

x = np.linspace(0, 10, 100)  # 100 values between 0 and 10
y = np.sin(x)                # sin(x) values

# Basic syntax:
# plt.plot(x, y, color=..., label=...)
plt.plot(x, y, color='blue', label='sin(x)')
plt.title('Line Plot')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True)               # Show grid lines
plt.legend()                 # Show legend
plt.show()
```

```python
# 📦 7. BOX PLOT

# Data: 3 datasets with different standard deviations
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# Basic syntax:
# plt.boxplot(data)
plt.boxplot(data)
plt.title('Box Plot')
plt.xlabel('Dataset')
plt.ylabel('Value')
plt.show()
```
