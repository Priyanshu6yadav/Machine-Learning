```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,2,3,4,5]
y = [0,9,8,7,6]
plt.plot(x,y)
```

```python
# Draw two points in the diagram, one at position (1, 3) and one in position (8, 10):
xpoints = np.array([1,8])
ypoints = np.array([3,10])

plt.plot(xpoints, ypoints, 'o')
plt.show()
```

### Markers

```python
# You can use the keyword argument marker to emphasize each point with a specified marker:
ypoints = np.array([3,8,1,10])

plt.plot(ypoints, marker = 'o')
plt.show()
```

```python
## mark each point with a star:
plt.plot(ypoints, marker = '>')
plt.show()
```

```python
# barplot
# Start the first wedge at 90 degrees:
y = np.array([35,25,25,15])
mylables = ['Apples','Bananas','Cherries','Dates']

plt.pie(y, labels = mylables, startangle = 90)
plt.show()
```

```python
y = np.array([35,25,25,15])
mylables = ['Apple','Bananas','Cherries','Dates']
myexplode = [0.1,0,0,0]

plt.pie(y, labels = mylables, explode = myexplode)
plt.show()
```

```python
### Shadow
y = np.array([35,25,25,15])
mylables = ['Apple','Bananas','Cherries','Dates']
myexplode = [0.1,0,0,0]

plt.pie(y, labels = mylables, explode = myexplode,shadow='True')
plt.legend()
plt.show()
```
