### SVM Kernals Indepth intutions And Practical Explanation

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(-5.0, 5.0, 100)
y = np.sqrt(10**2 - X**2)
y = np.hstack([y, -y])
X = np.hstack([X,-X])
```

```python
X1 = np.linspace(-5.0, 5.0, 100)
y1 = np.sqrt(5**2 - X1**2)
y1 = np.hstack([y1, -y1])
X1 = np.hstack([X1,-X1])
```

```python
plt.scatter(y,X)
plt.scatter(y1,X1)
```

```python
import pandas as pd
df1 = pd.DataFrame(np.vstack([y,X]).T,columns=['X1','X2'])
df1["Y"]=0
df2 = pd.DataFrame(np.vstack([y1,X1]).T,columns=['X1','X2'])
df2['Y']=1
df = pd.concat([df1,df2])
df.head(5)
```

```python
df.tail()
```

```python
##  Independent and dependent features
X = df.iloc[:, :2]
y = df.Y
```
