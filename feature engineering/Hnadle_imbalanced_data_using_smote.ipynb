### 1. SMOTE (Synthetic Minority Over-sampling Technique)

* Definition: A smarter up-sampling method that creates synthetic (new) minority samples instead of just duplicating.
* How: It interpolates between existing minority samples and their nearest neighbors to generate new, realistic data points.
* Benefit: Reduces overfitting compared to plain up-sampling.

```python
from sklearn.datasets import make_classification
```

```python
X,y=make_classification(n_samples=1000,n_redundant=0,n_features=2,n_clusters_per_class=1,
                    weights=[0.90],random_state=12)
# X is my independet features
# y is my dependent feature
```

```python
import pandas as pd
df1=pd.DataFrame(X,columns=['f1','f2'])
df2=pd.DataFrame(y,columns=['target'])
final_df=pd.concat([df1,df2],axis=1)
final_df.head()
```

```python
final_df['target'].value_counts()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(final_df['f1'],final_df['f2'],c=final_df['target'])
```

```python
from imblearn.over_sampling import SMOTE
```

```python
# transform the dataset
oversample=SMOTE()
X,y=oversample.fit_resample(final_df[['f1','f2']],final_df['target'])
```
