```python
# Convert publish_time to datetime
df['publish_time'] = pd.to_datetime(df['publish_time'])

# Convert trending_date to datetime (year.day.month format)
df['trending_date'] = pd.to_datetime(df['trending_date'], format='%y.%d.%m')
```

```python
# Convert publish_time to datetime and remove timezone
df['publish_time'] = pd.to_datetime(df['publish_time']).dt.tz_localize(None)

# Convert trending_date to datetime (year.day.month format)
df['trending_date'] = pd.to_datetime(df['trending_date'], format='%y.%d.%m')
```

```python
# Now subtraction works
df['days_to_trend'] = (df['trending_date'] - df['publish_time']).dt.days
```

```python
# Extract publish features
# Extract hour, weekday, and month from publish time
df['publish_hour'] = df['publish_time'].dt.hour
df['publish_day'] = df['publish_time'].dt.dayofweek   # Monday=0
df['publish_month'] = df['publish_time'].dt.month
```

```python
df = df.drop(columns=['publish_time'])
```

```python
# Text length features
df['title_length'] = df['title'].str.len()
```

```python
# Tags count
df['tags_count'] = df['tags'].apply(lambda x: len(x.split('|')))
```

```python
# Engagement ratios
df['likes_view_ratio'] = df['likes'] / (df['views'] + 1)
df['comments_view_ratio'] = df['comment_count'] / (df['views'] + 1)
```
